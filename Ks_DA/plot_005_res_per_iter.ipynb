```python
# # different dzeta factors
# n_iter = [0,1,3,5,7,3,3]
# rmse_smap_val = [0.12,0.09,0.06, 0, 0.06, 0, 0]
# r_smap_val =    [0.68,0.75,0.80, 0, 0.84, 0, 0]
# b_smap_val =    [-0.05,-0.03,-0.01, 0, 0.01, 0, 0]
# rmse_flx_val =  [0.85,0.85,0.85, 0, 0.84, 0, 0]
# r_flx_val =     [0.79,0.79,0.79, 0, 0.79, 0, 0]
# b_flx_val =     [-0.05,-0.02,0.03, 0, 0.09, 0, 0]

rmse_smap_val = []
r_smap_val =    []
b_smap_val =    []
rmse_flx_val =  []
r_flx_val =     []
b_flx_val =     []

rmse_smap_train = []
rmse_flx_train = []
r_smap_train =    []
b_smap_train =    []
r_flx_train =     []
b_flx_train =     []

b_flipped = np.array([True,True,True,False,True,True,True])

base_folder = '/p/scratch/cjibg36/kaandorp2/TSMP_results/eTSMP/'
folders = ['DA_eCLM_cordex_444x432_v14_1y_iter1','DA_eCLM_cordex_444x432_v14_1y_iter3',
           'DA_eCLM_cordex_444x432_v14_1y_iter5','DA_eCLM_cordex_444x432_v14_1y_iter7',
           'DA_eCLM_cordex_444x432_v14_1y_iter3_SMAPOnly','DA_eCLM_cordex_444x432_v14_1y_iter3_FLXOnly']

for i1,folder_ in enumerate(folders):
    data = load_dict_from_pickle(os.path.join(base_folder,folder_,'figures/03_correlations/stats_overall.pickle'))
    
    if i1 == 0: #add prior parameter results
        rmse_smap_val.append(data['SMAP']['val_prior']['rmse'])
        r_smap_val.append(data['SMAP']['val_prior']['R'])
        b_smap_val.append(-data['SMAP']['val_prior']['bias'])
        rmse_smap_train.append(data['SMAP']['train_prior']['rmse'])
        r_smap_train.append(data['SMAP']['train_prior']['R'])
        b_smap_train.append(-data['SMAP']['train_prior']['bias'])
        
        rmse_flx_val.append(data['FLX']['val_prior']['rmse'])
        r_flx_val.append(data['FLX']['val_prior']['R'])
        b_flx_val.append(-data['FLX']['val_prior']['bias'])
        rmse_flx_train.append(data['FLX']['train_prior']['rmse'])       
        r_flx_train.append(data['FLX']['train_prior']['R'])
        b_flx_train.append(-data['FLX']['train_prior']['bias'])
        
        print(data)
    rmse_smap_val.append(data['SMAP']['val_post']['rmse'])
    r_smap_val.append(data['SMAP']['val_post']['R'])
    b_smap_val.append(-data['SMAP']['val_post']['bias'])
    rmse_smap_train.append(data['SMAP']['train_post']['rmse'])
    r_smap_train.append(data['SMAP']['train_post']['R'])
    b_smap_train.append(-data['SMAP']['train_post']['bias'])
    
    rmse_flx_val.append(data['FLX']['val_post']['rmse'])
    r_flx_val.append(data['FLX']['val_post']['R'])
    b_flx_val.append(-data['FLX']['val_post']['bias'])
    rmse_flx_train.append(data['FLX']['train_post']['rmse'])     
    r_flx_train.append(data['FLX']['train_post']['R'])
    b_flx_train.append(-data['FLX']['train_post']['bias'])
    
    if i1 == 2:
        print(data)
```
