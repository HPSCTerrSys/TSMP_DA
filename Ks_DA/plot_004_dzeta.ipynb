```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# different dzeta factors
dzeta = [None,.2,.4,.6,.8]
r_smap = [0.688,0.803,0.827,0.829,0.846]
rmse_smap =  [1.22e-1,7.04e-2,6.10e-2,6.12e-2,5.58e-2]
b_smap  = -np.array([-0.059,-0.020,-0.007,-0.008,-0.005]) #change; +: wet bias, -: dry bias

r_flx = [0.773,0.769,0.782,0.771,0.747]
rmse_flx = [9.64e-1,9.61e-1,9.26e-1,9.58e-1,1.03]
b_flx = -np.array([-0.075,0.032,0.091,-0.001,-0.015]) 

```

```python
%matplotlib inline
all_labels = []
all_handles = []

cmap = plt.cm.tab10
colors = ['#1b9e77','#d95f02','#7570b3']

fig, ax = plt.subplots(figsize=(16,5))
fig.subplots_adjust(left=.2)
fig.subplots_adjust(right=.8)

ax.set_xlabel(r"$\zeta$") 
ax.set_ylabel('RMSE (SM)', color=colors[0],size=12) 
ax.tick_params(axis='y', labelcolor=colors[0], labelsize=10) 
ax.spines["left"].set_visible(True)
ax.yaxis.set_label_position('left')
ax.yaxis.set_ticks_position('left')
ax.plot(dzeta[1:], rmse_smap[1:], 'o-', color=colors[0],label='RMSE, SM')
h_,l_ = ax.get_legend_handles_labels()
all_labels.extend(l_)
all_handles.extend(h_)

ax_1 = ax.twinx() 
ax_1.yaxis.tick_left()
ax_1.set_ylabel('R (SM)', color=colors[1],size=12) 
ax_1.tick_params(axis='y', labelcolor=colors[1], labelsize=10) 
ax_1.plot(dzeta[1:], r_smap[1:], 'o-', color=cmap(1),label='R, SM')
ax_1.spines["left"].set_position(("axes", -0.08))    # Set second Y-axis 10% away from existing Y axis
ax_1.spines["left"].set_visible(True)
ax_1.yaxis.set_label_position('left')
ax_1.yaxis.set_ticks_position('left')
h_,l_ = ax_1.get_legend_handles_labels()
all_labels.extend(l_)
all_handles.extend(h_)

ax_2 = ax_1.twinx() 
ax_2.yaxis.tick_left()
ax_2.set_ylabel('bias (SM)', color=colors[2],size=12) 
ax_2.tick_params(axis='y', labelcolor=colors[2], labelsize=10) 
ax_2.plot(dzeta[1:], b_smap[1:], 'o-', color=colors[2],label='bias, SM')
ax_2.spines["left"].set_position(("axes", -0.16))    # Set second Y-axis 10% away from existing Y axis
ax_2.spines["left"].set_visible(True)
ax_2.yaxis.set_label_position('left')
ax_2.yaxis.set_ticks_position('left')
h_,l_ = ax_2.get_legend_handles_labels()
all_labels.extend(l_)
all_handles.extend(h_)

ax_3 = ax.twinx() 
ax_3.yaxis.tick_left()
ax_3.set_ylabel('RMSE (ET)', color=colors[0],size=12) 
ax_3.tick_params(axis='y', labelcolor=colors[0], labelsize=10) 
ax_3.plot(dzeta[1:], rmse_flx[1:], 'o--', color=colors[0],label='RMSE, ET')
ax_3.spines["right"].set_position(("axes", 1.))    # Set second Y-axis 10% away from existing Y axis
ax_3.spines["right"].set_visible(True)
ax_3.yaxis.set_label_position('right')
ax_3.yaxis.set_ticks_position('right')
h_,l_ = ax_3.get_legend_handles_labels()
all_labels.extend(l_)
all_handles.extend(h_)

ax_4 = ax.twinx() 
ax_4.yaxis.tick_left()
ax_4.set_ylabel('R (ET)', color=colors[1],size=12) 
ax_4.tick_params(axis='y', labelcolor=colors[1], labelsize=10) 
ax_4.plot(dzeta[1:], r_flx[1:], 'o--', color=colors[1],label='R, ET')
ax_4.spines["right"].set_position(("axes", 1.08))    # Set second Y-axis 10% away from existing Y axis
ax_4.spines["right"].set_visible(True)
ax_4.yaxis.set_label_position('right')
ax_4.yaxis.set_ticks_position('right')
h_,l_ = ax_4.get_legend_handles_labels()
all_labels.extend(l_)
all_handles.extend(h_)

ax_5 = ax.twinx() 
ax_5.yaxis.tick_left()
ax_5.set_ylabel('bias (ET)', color=colors[2],size=12) 
ax_5.tick_params(axis='y', labelcolor=colors[2], labelsize=10) 
ax_5.plot(dzeta[1:], b_flx[1:], 'o--', color=colors[2],label='bias, ET')
ax_5.spines["right"].set_position(("axes", 1.16))    # Set second Y-axis 10% away from existing Y axis
ax_5.spines["right"].set_visible(True)
ax_5.yaxis.set_label_position('right')
ax_5.yaxis.set_ticks_position('right')
h_,l_ = ax_5.get_legend_handles_labels()
all_labels.extend(l_)
all_handles.extend(h_)

ax.legend(handles=all_handles,labels=all_labels,loc='center right')
fig.savefig('figures/dzeta.png')
fig.savefig('figures/dzeta.pdf')
```
